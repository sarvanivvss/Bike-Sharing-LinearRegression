## Train and test split

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(Bike_df, train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
train.info()
```

## Re scaling feature using MinMax Scaler

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
# Apply scaler() to all the columns except the 'dummy' variables.
num_vars = ['Count','Humidity','windspeed','temp','atemp']

train[num_vars] = scaler.fit_transform(train[num_vars])
```

```python
train.head()
```

```python
train.describe()
```

```python
plt.figure(figsize = (20, 25))
sns.heatmap(train.corr(), annot = True, cmap="YlGnBu")
plt.show()
```

From the map, atemp and temp seems to be correlated to the target variable count very strongly. Since, not much can be stated about the other independent variables , hence we'll build a model using all the columns.

```python
y_train = train.pop('Count')
X_train = train
```

```python
y_train.shape
```

```python
X_train.shape
```

## Feature selection

```python
# Importing RFE and LinearRegression
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
# Running RFE with the output number of the variable equal to 15
lm = LinearRegression()
lm.fit(X_train, y_train)

rfe = RFE(estimator=lm, n_features_to_select=15)
rfe.fit(X_train, y_train)

```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
#creating a list of features selected by RFE

rfe_col = X_train.columns[rfe.support_]
rfe_col
```

```python
# feature which are chosen by RFE during feature selection( so un-supported columns)

X_train.columns[~rfe.support_]
```

Building model using statsmodel, for the detailed statistics

```python
# Creating X_test dataframe with RFE selected variables
X_train_rfe = X_train[rfe_col]
```

### Adding a constant variable

```python
import statsmodels.api as sm
X_train_rfe = sm.add_constant(X_train_rfe)
```

```python
X_train_rfe.head()
```

Running Linear Model

```python
lm_1 = sm.OLS(y_train,X_train_rfe).fit()
```

```python
lm_1.params
```

```python
lm_1.summary()
```

All p values <0.05

```python
# importing libraries to check the VIF values of the feature variables.

from statsmodels.stats.outliers_influence import variance_inflation_factor

# Create a dataframe that will contain the names of all the feature variables and their respective VIFs

vif = pd.DataFrame()
vif['Features'] = X_train_rfe.columns
vif['VIF'] = [variance_inflation_factor(X_train_rfe.values, i) for i in range(X_train_rfe.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

We generally want a VIF that is less than 5. So we need to drop Const.

```python
# dropping `const` column as the vif is > 5
X_train_rfe = X_train_rfe.drop(['const'], axis=1)
```

```python
# Calculate the VIFs for the new model again
vif = pd.DataFrame()
vif['Features'] = X_train_rfe.columns
vif['VIF'] = [variance_inflation_factor(X_train_rfe.values, i) for i in range(X_train_rfe.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
# dropping `hum` column as the vif is > 5
X_train_rfe = X_train_rfe.drop(['Humidity'], axis=1)
```

```python
# Calculate the VIFs for the new model again
vif = pd.DataFrame()
vif['Features'] = X_train_rfe.columns
vif['VIF'] = [variance_inflation_factor(X_train_rfe.values, i) for i in range(X_train_rfe.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
X_train_lm3 = sm.add_constant(X_train_rfe)

# Create a first fitted model
lr = sm.OLS(y_train, X_train_lm3).fit()

lr.summary()
```

```python
vif = pd.DataFrame()
vif['Features'] = X_train_rfe.columns
vif['VIF'] = [variance_inflation_factor(X_train_rfe.values, i) for i in range(X_train_rfe.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

we are not dropping temp as it has p value <0.05 and highly correlated with count

## Residual Analysis of the train data

```python
y_train_cnt = lr.predict(X_train_lm3)
```

```python
# Plot the histogram of the error terms
fig = plt.figure()
sns.distplot((y_train - y_train_cnt), bins = 20)
fig.suptitle('Error Terms', fontsize = 20)                  # Plot heading
plt.xlabel('Errors', fontsize = 18)                         # X-label
```

We can clearly see that the error terms are centred around 0 and follows a normal distribution, this is in accordance with the stated assumptions of linear regression

```python
# Plot the qq-plot of the error terms
sm.qqplot((y_train - y_train_cnt), fit=True, line='45')
plt.show()
```

Conclusion:
Here we see that most of the data points lie on the straight line , which indicates that the error terms are normally distributed .

## Making Predictions Using the Final Model

```python
num_vars = ['Count','Humidity','windspeed','temp','atemp']


test[num_vars] = scaler.transform(test[num_vars])
```

```python
test.describe()
```

```python
y_test = test.pop('Count')
X_test = test
```

```python
# Adding constant variable to test dataframe
X_test = sm.add_constant(X_test)
```

```python
test_col = X_train_lm3.columns
X_test=X_test[test_col[1:]]
# Adding constant variable to test dataframe
X_test = sm.add_constant(X_test)

X_test.info()
```

```python
# Making predictions using the final model

y_pred = lr.predict(X_test)
```

## Calculating the r-squared
R-squared is a goodness-of-fit measure for linear regression models. This statistic indicates the percentage of the variance in the dependent variable that the independent variables explain collectively. R-squared measures the strength of the relationship between your model and the dependent variable on a convenient 0  100% scale.

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

We can see that the r_squared on the test set is 0.8045 and the r-squared on the trained set 0.843 which is quiet reasonable and nearly equal, which means that whatever data the model was trained with, it has been almost able to apply those learnings in the test data.
